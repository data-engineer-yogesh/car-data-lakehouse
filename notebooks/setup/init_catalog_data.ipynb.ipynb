  Unity Catalog Design
name -> cardatalakehouse_cat

In Databricks, Unity Catalog is used for centralized data governance. The catalog structure can be created using SQL commands or via the UI by defining a catalog and schemas. Additionally, a volume can be created to store and manage the files required for the project.





First we list the existing catalogs.

```sql
ShOW CATALOGS;
```

Let's create the cardatalakehouse_cat

```sql

CREATE CATALOG IF NOT EXISTS cardatalakehouse_cat;

COMMENT ON CATALOG cardatalakehouse_cat IS'Catalog for cars lakehouse project';
```

Here, we verify the currently active catalog

```sql
SELECT current_catalog();
```

We need to select the catalog `cardatalakehouse_cat` for use.

```sql
USE CATALOG cardatalakehouse_cat;

SELECT current_catalog();
     
```

Retrieve more details about the cardatalakehouse_cat catalog


```sql
DESCRIBE CATALOG EXTENDED cardatalakehouse_cat;
```

We will create schemas aligned with the Medallion Architecture, including Bronze, Silver, and Gold layers

```sql
CREATE SCHEMA IF NOT EXISTS Bronze;

CREATE SCHEMA IF NOT EXISTS Silver;

CREATE SCHEMA IF NOT EXISTS Gold;

CREATE SCHEMA IF NOT EXISTS landing;
     
```

List the schemas we created.

```sql
SHOW SCHEMAS;
```

The landing schema will be used to create volumes that store raw or source data files before they are processed in the Bronze layer.


```sql
CREATE VOLUME If NOT EXISTS  cardatalakehouse_cat.landing.source_data;
```

UI : Please access the Unity Catalog and verify the setup for
- Catalog: cardatalakehouse_cat
- Schema: landing
- Volume: source_data

Upload all your source/raw data files into the source_data volume. Ensure only raw files are stored here, as this layer is used for initial data ingestion in the pipeline.
